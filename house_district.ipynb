```python
import pandas as pd
import numpy as np
import math
```

```python
house=pd.read_csv("https://dataverse.harvard.edu/api/access/datafile/:persistentId?persistentId=doi:10.7910/DVN/IG0UN2/8KAB8V", delimiter='\t')
```

```python
#This method pads all state fips with a 0 when they are less than 10, as well as handles null values.
def state_fips_code(x):
    l=[]
    for i in x:
        if i < 10:
            i = str(i)
            l.append("0"+i)
        else:
            i = str(i)
            l.append(i)
    return l
```

```python
#Appends the congressional district prefix to the state_fips and district ID
def district_fips(d, s):
    l=[]
    for i in range(0, d.size):
        if np.isnan(d.at[i]):
            l.append(str(d.at[i]))
        elif d.at[i] < 10:
            dis_code = "50000US" + s.at[i] + "0" + str(d.at[i])
            l.append(dis_code)
        else:
            dis_code = "50000US" + s.at[i] + str(d.at[i])
            l.append(dis_code)
    return l
```
