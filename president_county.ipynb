```python
import pandas as pd
import math
```

```python
presidents=pd.read_csv("https://dataverse.harvard.edu/api/access/datafile/:persistentId?persistentId=doi:10.7910/DVN/VOQCHQ/LK4XSK",delimiter='\t')
```

```python
def fips_code(x):
    l=[]
    for i in x:
        if i == None:
            l.append(str(i))
        elif math.isnan(i):
            l.append(str(i))
        elif i < 10000.0:
            i = str(int(i))
            l.append("05000US0"+i)
        else:
            i = str(int(i))
            l.append("05000US"+i)
    return l
```

```python
total=presidents.isnull().sum().sort_values(ascending=False)
percent=(presidents.isnull().sum()/presidents.isnull().count()).sort_values(ascending=False)
pd.concat([total,percent],keys=["Total","Percent"],axis=1)
```

```python
presidents[(presidents['FIPS'].isnull())&(presidents['state_po'].isnull())]
```

```python
presidents.drop("version",axis='columns',inplace=True)
presidents['FIPS']=fips_code(presidents['FIPS'])
presidents.loc[(presidents['county']=='Statewide writein')&(presidents['state']=='Connecticut'), 'state_po']='CT'
presidents.loc[(presidents['state']=='Maine')&(presidents['state_po'].isnull()), 'state_po']='ME'
presidents.loc[(presidents['state']=='Rhode Island')&(presidents['state_po'].isnull()), 'state_po']='RI'
presidents.loc[(presidents['state']=='Alaska')&(presidents['state_po'].isnull()), 'state_po']='AK'
presidents.loc[(presidents['party'].isnull()), 'party']='Other'
```

```python
total=presidents.isnull().sum().sort_values(ascending=False)
percent=(presidents.isnull().sum()/presidents.isnull().count()).sort_values(ascending=False)
pd.concat([total,percent],keys=["Total","Percent"],axis=1)
```
